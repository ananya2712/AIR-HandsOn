# PREPROCESSING

```python
d = dfinal["data"].tolist() # convert documents to a list 
```

```python
! pip install tweet-preprocessor
clean = []
import preprocessor as p

def preproc(tweet):
  clean_tweet = p.clean(tweet)
  return clean_tweet

# Cleaning
for i in d:
  clean.append(preproc(i))
print(clean[0])
```

```python
# REMOVAL OF STOP WORDS
res = [] # resultant list of cleaned strings

for s in clean:
  lst_string = [s][0].split()
  
  no_stpwords_string="" #new string created 
  for i in lst_string:
    if not i in stop_words:
        no_stpwords_string += i + ' '
  no_stpwords_string = no_stpwords_string[:-1]
  res.append(no_stpwords_string)

print(res[0]) # example cleaned string
```

```python
# REMOVE PUNCTUATION
def remove_punctuation(words):
  new_words = []
  for word in words:
    new_word = re.sub(r'[^\w\s]', ' ', (word))
    if new_word != '':
       new_words.append(new_word)
  return new_words

new = remove_punctuation(res)
print(new[0])
```

```python
# REMOVE SPACES AND DIGITS 
data_final = []

def spaces_digits(tweet):
    tweet = re.sub('\s+', ' ', tweet)  # remove multi spacing
    tweet = re.sub('([0-9]+)', '', tweet)  # remove numbers
    return tweet

for i in new:
  data_final.append(spaces_digits(i))

print(data_final[0])
```

```python
# TOKENISE THE DOCUMENT
from nltk.tokenize import regexp_tokenize
  
res_fin = [sub.split() for sub in data_final] #creates a list of lists for all documents tokenised
print(res_fin[0])
```

```python
# CONVERT BACK TO DATAFRAME
df = pd.DataFrame(columns=['data'])
for i in res_fin:
  temp = {'data':i}
  df = df.append(temp,ignore_index = True)
df
```

INVERTED INDEX USING HASHING FUNCTION

```python
inverted_index = dict()
docs = len(res_fin)

for i in range(docs):
  for j in res_fin[i]:
    indices = []
    if j not in inverted_index:
      for k in range(docs):
        if j in res_fin[k]:
          indices.append(k + 1)
      inverted_index[j] = indices[:]
```

BINARY SEARCH TREE FUNCTIONS

```python
# tree node

class Node:
  def __init__(self, token, index_list):
    self.left = None
    self.right = None
    self.token = token
    self.doc_ids = index_list


def tree_search(root, word):
  
  # return [] if word not found in tree else return [indices of docs]
  while root:
    if word == root.token:
      return root.doc_ids
    if word[0] <= root.token[0]:
      root = root.left
    elif word[0] > root.token[0]:
      root = root.right
  return [] 


def tree_insert(root, word, index_list):
  newnode = Node(word, index_list)
  if not root:
    root = newnode

  temp = root
  prev = None

  while temp:
    prev = temp
    if word[0] <= temp.token[0]:
      temp = temp.left
    else:
      temp = temp.right
  
  if word[0] <= prev.token[0]:
    prev.left = newnode
  else:
    prev.right = newnode

def inorder(root):
  if root:
    try:
      inorder(root.left)
      print(root.token, root.doc_ids)
      inorder(root.right)
    except RecursionError:
      return
```

```python
docs = len(res_fin)
root = Node('__initialnode__', [None])

for i in range(docs):
  for j in res_fin[i]:
    indices = []
    temp1 = root
    temp2 = root
    if not tree_search(temp1, j):
      for k in range(docs):
        if j in res_fin[k]:
          indices.append(k + 1)
      tree_insert(temp2, j, indices)
inorder(root)     
```
